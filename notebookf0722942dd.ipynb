```python
### Train/Test split
from sklearn.model_selection import train_test_split
features = ["bedrooms", "bathrooms", "sqft_living","floors","waterfront","yr_built","yr_renovated", "view", "sqft_above", "sqft_basement"]

X = house_data[features].copy()
y = house_data.price

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.6, test_size = 0.4, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

model1 = RandomForestRegressor(n_estimators = 100, criterion='squared_error')
model2 = RandomForestRegressor(n_estimators = 100, criterion='squared_error', max_depth = 10)
model3 = LinearRegression()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import log_loss

model1.fit(X_train, y_train)
predicted_prices = model1.predict(X_test)
print("MAE : " , mean_absolute_error(y_test, predicted_prices))
```

```python
model2.fit(X_train, y_train)
predicted_prices2 = model2.predict(X_test)
mean_absolute_error(y_test, predicted_prices2)
```

```python
model3.fit(X_train, y_train)
predicted = model3.predict(X_test)
mean_absolute_error(y_test,predicted)
```
